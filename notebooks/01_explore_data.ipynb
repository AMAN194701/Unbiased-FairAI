```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
columns = ['Age','Job_type','Final_wt','Education_level','Education_Yr',
           'Marital_status','Job_role','Family_role','Race',
           'Gender','Capital_gain','Capital_loss','Weekly_hrs',
           'Country','Income']
```

```python
import os
df = pd.read_csv('data/adult.data', names=columns, na_values='?', skipinitialspace=True)
print(df.shape)
df.head()
```

```python
# checking missing values 
missing_values = df.isna().sum()
print(missing_values)
```

```python
print("Shape before : ", df.shape)
```

```python
df_clean = df.dropna()
```

```python
print("Shape before cleaning : ",df.shape)
print("Shape after cleaning :", df_clean.shape )
```

```python
print("Rows remove : ",df.shape[0] -  df_clean.shape[0])
```

```python
print("No. of rows and col in clean dataset : ",df_clean.shape)
```

```python
df= df_clean
```

```python
df.head()
```

```python
df["Income"].value_counts()
```

```python
df["Income"].value_counts(normalize=True)*100
```

We can say data is imbalanced since nearly 75% are earning <=50k which can lead to biased model predictions toward the majority class.

```python
df["Gender"].value_counts()
```

```python
pd.crosstab(df["Gender"], df["Income"],normalize="index")*100
```

The dataset exhibits strong gender bias, where only ~11% of females earn >50K compared to ~31% of males, indicating significant disparity in outcomes.

```python
pd.crosstab(df["Race"], df["Income"], normalize="index")*100
```

The dataset shows significant racial bias, where Asian-Pac-Islander (~27.7%) and White (~26.3%) groups have a much higher proportion of individuals earning >50K compared to groups like Black (~13%), Amer-Indian-Eskimo (~11.8%), and Other (~9%).
